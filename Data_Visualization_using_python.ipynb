```python
Scored_above_zero = df['Score'] > '0-0'
```

```python
df[Scored_above_zero]
```

```python
Scored_by_brazil = df['home_team'] == 'Brazil'
```

```python
Scored_by_brazil
```

```python
brazildf =df[Scored_above_zero & Scored_by_brazil].sort_values (by ='home_score',ascending=True)
```

```python
brazildf
```

```python
from matplotlib import pyplot as plt
```

```python
brazildf.plot(x = 'home_score',y= 'away_score',kind='scatter')
plt.xlabel('Home score')
plt.ylabel('away score')
plt.legend('Takeaway')
plt.title('Brazil_Matches')
plt.show()
```

```python
brazildf.plot(x='Year',y='home_score',kind='hexbin')
plt.xlabel('Year')
plt.ylabel('Home score')
plt.legend('Takeaway')
plt.title('Brazil')
plt.show()
```
