```python
#Terms in machine learning

#1.Collection    : Data used for training,validating and testing the machine learning model.

#2.Features(x)   : Independent variables or input variable -used to predict the target variable.

#3.Target (y)    : Dependent variable - output -we want to predict Ex:- predicting number of bed rooms.

#4.Training Data : Subset of Dataset used to Train the model.

#5.Testing Data  : Subset of Data set to validate the trained model.

#6.Model         : Mathematical Representation of between features and Target.

#7.Supervised
   learning (SL) : Training the data based on label outcome or data\ the model learns from known/labelled data.

#8.Regression    : Method to predict continuous values from a model - supervised learning.

#9.Overfitting   : when a model learns from traning data too well ,includes
                   noise -perfrom poorly new data.

#10.Underfitting : Model is too simple and does not capture patterns in data .


## categories in Machine Learning:

# 1. Supervised learning  :
#     Model learns from labelled data -I/O pairs
#     Examples : linear regression, decision tress, Random forest,SVM
#     Best use case so far: House price prediction based on previous/historical data.

# 2.Unsupervised Learning :
#     Model finds new patterns in unlabel data without pre-defined data / labels.
#     Examples: Clustering (k means ),PCA
#     Use case : customer segmentation in marketing

# 3.Reinforcement learning:
#     Model learns by interacting with environment and receiveing the rewards or penalities .
#     Examples: Deep learning,AlphaGo...
#     Best use cases : self car driving vechicels,self learning AI bots,SLAM
```

```python
# Examples : predicting the house price - random dataset - generated

# Step:
# 1. Generating a random dataset
# 2. Train a ML model - linear regression
# 3. Evaluate the Model
# 4. Make a prediction
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python



# step 1 : Generate a random Dataset
np.random.seed(42) # used to set same sequence of values from the random package.
numSamples = 100
size = np.random.randint(800,5000,numSamples) #create numSamples number of samples from (800,5000)
bedrooms = np.random.randint(1,6, numSamples)
washrooms = np.random.randint(1,8, numSamples)
localityscore = np.random.randint(1,10, numSamples)

#Generate House prices with randomness
houseprice = (size*3000) + (bedrooms * 50000) + (washrooms * 30000)+(localityscore * 70000)+ np.random. randint(-50000 , 50000, numSamples)

#create a dataframe
df = pd.DataFrame(
     {
        'Size' : size,
        'Bedrooms' : bedrooms,
        'Washrooms' : washrooms,
        'Locality score' : localityscore,
        'Price' : houseprice
     }
)
```

```python
df
```

```python
# Step 2 : Train a ML model - linear regression

#packages(installation) regression :
#pip install scikit-learn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X= df[['Size','Bedrooms','Washrooms','Locality score']]
Y= df['Price']
```

```python
X_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state =42)

model=LinearRegression()
model.fit(X_train,y_train)
```

```python
# Step 3 : Evaluate the Model
#packages Requirements:
#sklearn.matrices

from sklearn.metrics import mean_absolute_error,mean_squared_error

y_pred = model.predict(x_test)

MAE = mean_absolute_error(y_test,y_pred)
MSE = mean_squared_error(y_test,y_pred)
RMSE = np.sqrt(MSE)

print(f'Mean Absolute Error :{MAE}')
print(f'Mean Squareed Error :{MSE}')
print(f'Root Mean Squareed Error :{RMSE}')
```

```python
#Step 4 : Make predictions
newHouse = np.array([[500,1,1,1]])
predictedprice= model.predict(newHouse)
print(f'the predicted price of new houses is {predictedprice[0]}')
```
